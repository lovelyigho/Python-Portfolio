```python
# Create a new colunmn that sums all years for each country
df['Total_Inflow'] = df.iloc[:, 1:].sum(axis=1)

# Sort by total inflow, descending
top10 = df.sort_values(by='Total_Inflow', ascending=False).head(10)

# Show the results
top10[['Country', 'Total_Inflow']]

# iloc[:, 1:] means â€œall columns except Countryâ€.

# .sum(axis=1) means sum across the row (yearly values).

# .sort_values() arranges from highest to lowest inflow.
```

```python
df = df[df['Country'] != 'World']
```

```python
df['Total_Inflow'] = df.iloc[:, 1:].sum(axis=1)
top10 = df.sort_values(by='Total_Inflow', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.barh(top10['Country'], top10['Total_Inflow'], color = 'skyblue')
plt.gca().invert_yaxis() # puts the highest bar at the top
plt.title('Top 10 Countries by Total Remittance Inflow (1970-2017)')
plt.xlabel('Country')
plt.ylabel('Total Inflow (US$ million)')
plt.show()
```

```python
# Filter for Nigeria
nigeria = df[df['Country'] == 'Nigeria']

# Remove the country column and transpose so years become rows
nigeria_trend = nigeria.iloc[:, 1:-1].T # Meaning everything except country and Total_Inflow
nigeria_trend.columns = ['Nigeria']

# Convert index to numeric (This fixes the slicing problem)
nigeria_trend.index = nigeria_trend.index.astype(int)

# Focus on recent years
nigeria_recent = nigeria_trend.loc[2000:2017]

# Plot the line chart
plt.figure(figsize=(10,5))
plt.plot(nigeria_recent.index, nigeria_recent['Nigeria'], marker='o', color='green')
plt.title('Nigeria â€“ Remittance Inflow Trend (2000â€“2017)')
plt.xlabel('Year')
plt.ylabel('Inflow (US$ Million)')
plt.grid(True)
plt.xticks(range(2000, 2018, 2)) # Label every 2 years
plt.show()
```

```python
countries = ['Nigeria', 'India', 'Mexico']
subset = df[df['Country'].isin(countries)]

# Transpose for plotting
subset_trend = subset.set_index('Country').T.iloc[: -1] # Exclude Total_Inflow
subset_trend.plot(figsize=(10, 6), linewidth=2)

plt.title('Remittance Inflows Comparison (1970-2017)')
plt.xlabel('Year')
plt.ylabel('US$ million')
plt.legend(title='Country')
plt.grid(True)
plt.show()
```

```python
# Sum across all countries for each year
yearly_totals = df.iloc[:, 1: -1].sum()

# Find the year with the max total
highest_year = yearly_totals.idxmax()
max_value = yearly_totals.max()

print(f'The year with the highest total global remittance inflow is {highest_year} with {max_value:,.2f} US$ million.')
```

## ðŸ“Š Summary of Findings

- **Top Countries:** From 1970â€“2017, India, China, and Mexico consistently recorded the highest remittance inflows globally, aligning with their large diaspora populations.

- **Nigeriaâ€™s Growth:** Nigeriaâ€™s remittance inflows remained under 2 billion dollars until around 2004, after which inflows surged rapidly, reaching over $20 billion annually by 2010 and remaining stable till 2017


- **Global Peak Year:** The year 2014 recorded the highest total global remittance inflow at approximately **$597.7 billion USD**, after which inflows plateaued.

- **Trend Insight:** Remittance growth in developing economies such as Nigeria often mirrors improvements in **financial inclusion**, **migration networks**, and **digital transfer options** (e.g., mobile money, fintech platforms).
